## 3. 자료형

Python에서 자주 쓰는 기본 자료형입니다.

```python
text = "문자열"
number = 10
price = 3.14
is_ready = True

print(type(text))
print(type(number))
print(type(price))
print(type(is_ready))
```

## 4. 리스트

리스트는 여러 값을 순서대로 담는 자료형입니다.

```python
tools = ["Host", "Client", "Server"]

print(tools)
print("첫 번째:", tools[0])
print("개수:", len(tools))
```

## 5. 딕셔너리

딕셔너리는 이름표(key)와 값(value)을 함께 저장합니다. JSON과 모양이 비슷해서 MCP 실습에서 자주 보게 됩니다.

```python
train_result = {
    "epoch": 0.12,
    "global_step": 500,
    "loss": 0.3358,
    "learning_rate": 0.0000188
}

print("loss:", train_result["loss"])
print("step:", train_result["global_step"])
```

## 6. 조건문

`if`는 조건에 따라 실행할 코드를 고를 때 사용합니다.

```python
loss = train_result["loss"]

if loss < 0.5:
    print("학습이 잘 진행되고 있습니다.")
else:
    print("loss가 높습니다. 확인이 필요합니다.")
```

## 7. 반복문

`for`는 리스트나 딕셔너리의 값을 하나씩 꺼내 처리할 때 사용합니다.

```python
for tool in tools:
    print("MCP 구성요소:", tool)

print("---")

for key, value in train_result.items():
    print(key, "=", value)
```

## 8. 함수

함수는 자주 쓰는 코드를 이름 붙여 재사용하는 방법입니다.

```python
def summarize_result(result):
    loss = result["loss"]
    step = result["global_step"]
    return f"현재 {step} step까지 학습했고, loss는 {loss}입니다."

summary = summarize_result(train_result)
print(summary)
```

## 9. 파일 저장과 읽기

MCP는 파일을 읽거나 쓰는 작업을 도구로 만들 수 있습니다. 먼저 Python으로 JSON 파일을 저장하고 읽어봅니다.

```python
import json

with open("sample_train_result.json", "w", encoding="utf-8") as f:
    json.dump(train_result, f, ensure_ascii=False, indent=2)

with open("sample_train_result.json", "r", encoding="utf-8") as f:
    loaded_result = json.load(f)

print(loaded_result)
```

## 10. MCP Tool처럼 생각하기

아래 함수는 MCP Tool로 바꾸기 쉬운 형태입니다.

입력으로 파일 경로를 받고, 결과를 딕셔너리로 돌려줍니다.

```python
def read_experiment_result(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return json.load(f)

result = read_experiment_result("sample_train_result.json")
print("파일에서 읽은 loss:", result["loss"])
```
